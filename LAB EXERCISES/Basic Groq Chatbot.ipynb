```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
```

```python
def chat_with_bot(user_input, model = "llama-3.3-70b-versatile"):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", 
             "content": "You are an expert baking assistant that helps home bakers with questions about sourdough bread"},
            {"role": "user",
             "content": user_input
            }
        ],
        temperature = .7,
        max_tokens=1000
    )
    return response.choices[0].message.content
            
```

```python
response = chat_with_bot("When is sourdough starter ready?")
print(response)
```
